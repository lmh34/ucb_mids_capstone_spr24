```python
## Libraries
import pandas as pd
from functools import reduce
```

```python
# Read each CSV file into a DataFrame
# Import NIBRS System Data
# 2021 and 2022 Data for California

df_NIBRS_CA_INCIDENT_2021 = pd.read_csv('CA-2021/NIBRS_incident.csv', usecols=['incident_date', 'incident_id', 'agency_id'])
df_NIBRS_CA_Agencies_2021 = pd.read_csv('CA-2021/agencies.csv', usecols=['county_name', 'agency_id','state_abbr'])
df_NIBRS_CA_Offense_2021 = pd.read_csv('CA-2021/NIBRS_offense.csv', usecols=['offense_id', 'incident_id', 'offense_code'])
df_NIBRS_CA_Offense_Type_2021 = pd.read_csv('CA-2021/NIBRS_offense_type.csv', usecols=['offense_name', 'offense_code'])
df_NIBRS_CA_INCIDENT_2022 = pd.read_csv('CA-2022/NIBRS_incident.csv', usecols=['incident_date', 'incident_id', 'agency_id'])
df_NIBRS_CA_Agencies_2022 = pd.read_csv('CA-2022/agencies.csv', usecols=['county_name', 'agency_id','state_abbr'])
df_NIBRS_CA_Offense_2022 = pd.read_csv('CA-2022/NIBRS_offense.csv', usecols=['offense_id', 'incident_id', 'offense_code'])
df_NIBRS_CA_Offense_Type_2022 = pd.read_csv('CA-2022/NIBRS_offense_type.csv', usecols=['offense_name', 'offense_code'])

# 2021 and 2022 Data for Texas
df_NIBRS_TX_INCIDENT_2021 = pd.read_csv('TX-2021/NIBRS_incident.csv', usecols=['incident_date', 'incident_id', 'agency_id'])
df_NIBRS_TX_Agencies_2021 = pd.read_csv('TX-2021/agencies.csv', usecols=['county_name', 'agency_id','state_abbr'])
df_NIBRS_TX_Offense_2021 = pd.read_csv('TX-2021/NIBRS_offense.csv', usecols=['offense_id', 'incident_id', 'offense_code'])
df_NIBRS_TX_Offense_Type_2021 = pd.read_csv('TX-2021/NIBRS_offense_type.csv', usecols=['offense_name', 'offense_code'])
df_NIBRS_TX_INCIDENT_2022 = pd.read_csv('TX-2022/NIBRS_incident.csv', usecols=['incident_date', 'incident_id', 'agency_id'])
df_NIBRS_TX_Agencies_2022 = pd.read_csv('TX-2022/agencies.csv', usecols=['county_name', 'agency_id','state_abbr'])
df_NIBRS_TX_Offense_2022 = pd.read_csv('TX-2022/NIBRS_offense.csv', usecols=['offense_id', 'incident_id', 'offense_code'])
df_NIBRS_TX_Offense_Type_2022 = pd.read_csv('TX-2022/NIBRS_offense_type.csv', usecols=['offense_name', 'offense_code'])

# 2021 and 2022 Data for New York
df_NIBRS_NY_INCIDENT_2021 = pd.read_csv('NY-2021/NIBRS_incident.csv',usecols=['incident_date', 'incident_id', 'agency_id'])
df_NIBRS_NY_Agencies_2021 = pd.read_csv('NY-2021/agencies.csv', usecols=['county_name', 'agency_id','state_abbr'])
df_NIBRS_NY_Offense_2021 = pd.read_csv('NY-2021/NIBRS_offense.csv', usecols=['offense_id', 'incident_id', 'offense_code'])
df_NIBRS_NY_Offense_Type_2021 = pd.read_csv('NY-2021/NIBRS_offense_type.csv', usecols=['offense_name', 'offense_code'])
df_NIBRS_NY_INCIDENT_2022 = pd.read_csv('NY-2022/NIBRS_incident.csv', usecols=['incident_date', 'incident_id', 'agency_id'])
df_NIBRS_NY_Agencies_2022 = pd.read_csv('NY-2022/agencies.csv', usecols=['county_name', 'agency_id','state_abbr'])
df_NIBRS_NY_Offense_2022 = pd.read_csv('NY-2022/NIBRS_offense.csv', usecols=['offense_id', 'incident_id', 'offense_code'])
df_NIBRS_NY_Offense_Type_2022 = pd.read_csv('NY-2022/NIBRS_offense_type.csv', usecols=['offense_name', 'offense_code'])

# Concatenate all years into one
df_NIBRS_INCIDENT = pd.concat([df_NIBRS_CA_INCIDENT_2021, df_NIBRS_CA_INCIDENT_2022, df_NIBRS_TX_INCIDENT_2021, 
                                  df_NIBRS_TX_INCIDENT_2022, df_NIBRS_NY_INCIDENT_2021, df_NIBRS_NY_INCIDENT_2021])
df_NIBRS_Agencies = pd.concat([df_NIBRS_CA_Agencies_2021, df_NIBRS_CA_Agencies_2022, df_NIBRS_TX_Agencies_2021, 
                                  df_NIBRS_TX_Agencies_2022, df_NIBRS_NY_Agencies_2021, df_NIBRS_NY_Agencies_2022])
df_NIBRS_Offense =  pd.concat([df_NIBRS_CA_Offense_2021,  df_NIBRS_CA_Offense_2022, df_NIBRS_TX_Offense_2021,
                                 df_NIBRS_TX_Offense_2022, df_NIBRS_NY_Offense_2021, df_NIBRS_NY_Offense_2022])
df_NIBRS_Offense_Type = pd.concat([df_NIBRS_CA_Offense_Type_2021, df_NIBRS_CA_Offense_Type_2022, df_NIBRS_TX_Offense_Type_2021,
                                     df_NIBRS_TX_Offense_Type_2022, df_NIBRS_NY_Offense_Type_2021, df_NIBRS_NY_Offense_Type_2022])

del df_NIBRS_CA_INCIDENT_2021
del df_NIBRS_CA_Agencies_2021
del df_NIBRS_CA_Offense_2021
del df_NIBRS_CA_Offense_Type_2021
del df_NIBRS_CA_INCIDENT_2022 
del df_NIBRS_CA_Agencies_2022 
del df_NIBRS_CA_Offense_2022 
del df_NIBRS_CA_Offense_Type_2022 

del df_NIBRS_TX_INCIDENT_2021
del df_NIBRS_TX_Agencies_2021
del df_NIBRS_TX_Offense_2021
del df_NIBRS_TX_Offense_Type_2021
del df_NIBRS_TX_INCIDENT_2022 
del df_NIBRS_TX_Agencies_2022 
del df_NIBRS_TX_Offense_2022 
del df_NIBRS_TX_Offense_Type_2022 

del df_NIBRS_NY_INCIDENT_2021
del df_NIBRS_NY_Agencies_2021
del df_NIBRS_NY_Offense_2021
del df_NIBRS_NY_Offense_Type_2021
del df_NIBRS_NY_INCIDENT_2022 
del df_NIBRS_NY_Agencies_2022 
del df_NIBRS_NY_Offense_2022 
del df_NIBRS_NY_Offense_Type_2022 
```

```python
#Merge NIBRS data into one 
# First merge: Merge df1 and df2 on "agency_id"
merged_NIBRS_1 = pd.merge(df_NIBRS_INCIDENT, df_NIBRS_Agencies, on='agency_id', how='left')

# Second merge: Merge the result of the first merge with Offense on "Incident ID"
merged_NIBRS_2 = pd.merge(merged_NIBRS_1, df_NIBRS_Offense, on='incident_id', how='left')

# Third merge: Merge the result of the second merge with Offense Type on "Offense_code"
merged_NIBRS_Final = pd.merge(merged_NIBRS_2, df_NIBRS_Offense_Type, on='offense_code', how='left')

del df_NIBRS_INCIDENT
del df_NIBRS_Agencies
del df_NIBRS_Offense
del df_NIBRS_Offense_Type
del merged_NIBRS_1
del merged_NIBRS_2
```

```python
print(merged_NIBRS_Final.info())
```

```python
columns_to_convert = ['county_name', 'offense_name', 'state_abbr']  # Add any other string columns
for column in columns_to_convert:
    merged_NIBRS_Final[column] = merged_NIBRS_Final[column].astype('category')

# Only keep the features needed for modeling or EDA
merged_NIBRS_Final = merged_NIBRS_Final[['state_abbr','county_name', 'incident_id', 'offense_name', 'incident_date']]

# Explicitly specify the date format if pandas fails to automatically recognize it
merged_NIBRS_Final['incident_date'] = pd.to_datetime(merged_NIBRS_Final['incident_date'], format='%Y/%m/%d')
```

```python
# Filter the DataFrame for rows where 'county_name' is 'ALAMEDA'
alameda_records = merged_NIBRS_Final[merged_NIBRS_Final['county_name'] == 'ALAMEDA']

# Find distinct values in the 'st_abbr' column of the filtered DataFrame
distinct_st_abbr = alameda_records['state_abbr'].unique()

distinct_st_abbr
```

```python

# Create a new column with dates formatted as "YYYYMM"
merged_NIBRS_Final['PERIOD'] = merged_NIBRS_Final['incident_date'].dt.strftime('%Y%m')

# Filter for 'Drugs' offenses and group by 'county_name', 'PERIOD', and state
merged_NIBRS_Final = merged_NIBRS_Final[merged_NIBRS_Final['offense_name'] == 'Drug/Narcotic Violations'].groupby(['county_name', 'PERIOD', 'state_abbr'])['incident_id'].nunique().reset_index(name='Drugs_Offense_Count')

#Capitalize all letters so all datasets match and can be joined correctly
merged_NIBRS_Final['county_name'] = merged_NIBRS_Final['county_name'].str.upper()

# Rename columns to match other data foundations
merged_NIBRS_Final = merged_NIBRS_Final.rename(columns={'state_abbr': 'ST_ABBREV'})

# Display the result
merged_NIBRS_Final
```

```python
# Import CDC Drug overdose data by county
df_CDC_Overdose = pd.read_csv('VSRR_Provisional_County-Level_Drug_Overdose_Death_Counts.csv')
# Filter the DataFrame to keep only records where "St_Abbrev" equals "CA"
df_CDC_Overdose = df_CDC_Overdose[(df_CDC_Overdose['ST_ABBREV'].isin(['CA', 'TX', 'NY'])) & (df_CDC_Overdose['Year'] > 2020)]

# Explicitly specify the date format if pandas fails to automatically recognize it
df_CDC_Overdose['MonthEndingDate'] = pd.to_datetime(df_CDC_Overdose['MonthEndingDate'], format='%m/%d/%Y')
# Create a new column with dates formatted as "YYYYMM"
df_CDC_Overdose['PERIOD'] = df_CDC_Overdose['MonthEndingDate'].dt.strftime('%Y%m')

# Only keep the features that will be used in the model or EDA
df_CDC_Overdose = df_CDC_Overdose[['COUNTYNAME', 'Provisional Drug Overdose Deaths', 
                                         'PERIOD', 'ST_ABBREV']]
```

```python
#Resort by county name and period
df_CDC_Overdose.sort_values(by=['COUNTYNAME', 'PERIOD'], ascending=[True, False])
df_CDC_Overdose.reset_index(drop=True, inplace=True)

# Rename columns to match other data foundations
df_CDC_Overdose = df_CDC_Overdose.rename(columns={'COUNTYNAME': 'county_name'})

#Capitalize all letters so all datasets match and can be joined correctly
df_CDC_Overdose['county_name'] = df_CDC_Overdose['county_name'].str.upper()

df_CDC_Overdose['PERIOD'] = df_CDC_Overdose['PERIOD'].astype(str)

df_CDC_Overdose
```

```python
# Import CDC Drug overdose data by county
df_Border_Patrol = pd.read_csv('nationwide-drugs-fy21-fy24-dec.csv')
# Filter the DataFrame to keep only records where "drug_type" contains "fentanyl" (case-insensitive)
df_BP_Fentanyl = df_Border_Patrol[df_Border_Patrol['Drug Type'].str.contains('fentanyl', case=False, na=False)]
# Filter the DataFrame to keep only records in California+
#df_BP_Fentanyl = df_BP_Fentanyl[df_BP_Fentanyl['Area of Responsibility'].str.contains('San Diego', case=False, na=False)]
# Remove any records from 2024
df_BP_Fentanyl = df_BP_Fentanyl[df_BP_Fentanyl["FY"] != "2024 (FYTD)"]

#Create a month mapping to be able to convert the date into a Period
month_mapping = {
    'JAN': '01', 'FEB': '02', 'MAR': '03', 'APR': '04',
    'MAY': '05', 'JUN': '06', 'JUL': '07', 'AUG': '08',
    'SEP': '09', 'OCT': '10', 'NOV': '11', 'DEC': '12'
}

# Convert 'Month' to two-digit format using the mapping
df_BP_Fentanyl['Month_MM'] = df_BP_Fentanyl['Month (abbv)'].map(month_mapping)

# Concatenate 'Year' and the new 'Month_MM' to form the six-digit code
# Ensure 'Year' is in string format before concatenation
df_BP_Fentanyl['PERIOD'] = df_BP_Fentanyl['FY'].astype(str) + df_BP_Fentanyl['Month_MM']

# Group by columns to get total for county and period
# Group by specified columns and aggregate
df_BP_Fentanyl = df_BP_Fentanyl.groupby(['FY', 'Month (abbv)', 'Month_MM', 'PERIOD'], as_index=False).agg({
    'Count of Event': 'sum',
    'Sum Qty (lbs)': 'sum'
})

# Only bring in the features needed for EDA or modeling
df_BP_Fentanyl = df_BP_Fentanyl[['PERIOD', 'Count of Event', 'Sum Qty (lbs)','FY']]

# Rename columns to match other data foundations
df_BP_Fentanyl = df_BP_Fentanyl.rename(columns={'Count of Event': 'BP_Count', 'Sum Qty (lbs)': 'BP_Weight', 'FY': 'YEAR'})

#Display the values sorted by Period
df_BP_Fentanyl
```

```python
#Import Census Population data
df_Census_CA = pd.read_csv('CENSUS/cc-est2022-agesex-06.csv')
df_Census_NY = pd.read_csv('CENSUS/cc-est2022-agesex-36.csv')
df_Census_TX = pd.read_csv('CENSUS/cc-est2022-agesex-48.csv')

# Concatenate all states into one
df_Census = pd.concat([df_Census_CA, df_Census_TX, df_Census_NY])

#delete datasets that will no longer be used
del df_Census_CA
del df_Census_TX
del df_Census_NY

# Rename columns to match other data foundations
df_Census = df_Census.rename(columns={'CTYNAME': 'county_name', 
                                     'STNAME': 'ST_ABBREV'})

#Capitalize all letters so all datasets match and can be joined correctly
df_Census['county_name'] = df_Census['county_name'].str.upper()
df_Census['YEAR'] = df_Census['YEAR'].astype(str)

df_Census
```

```python
#Merge all separate data foundations into one final df
# Merge df_CDC_Overdose with the subset of merged_NIBRS_Final
df_Supplemental_Data = pd.merge(df_CDC_Overdose, merged_NIBRS_Final, on=['PERIOD', 'county_name', 'ST_ABBREV'], how='left')
#Merge the border patrol data with the main df
df_Supplemental_Data = pd.merge(df_Supplemental_Data, df_BP_Fentanyl, on=['PERIOD'], how='left' )
#Merge the Population and Age data with the main df
df_Supplemental_Data = pd.merge(df_Supplemental_Data, df_Census, on=['county_name', 'ST_ABBREV', 'YEAR'], how='left' )
```

```python
#Use population 16 and over to calculate drug overdose rate per county
#Overdose_Rate_X100K
df_Supplemental_Data['Overdose_Rate_X100K'] = 100000 * (df_Supplemental_Data['Provisional Drug Overdose Deaths']/df_Supplemental_Data['AGE16PLUS_TOT'])

# Remove the overdose rate column from the list
cols = [col for col in df_Supplemental_Data.columns if col != 'Overdose_Rate_X100K']  

# Insert overdose rate column at the third position (index 2 because of zero-based indexing)
cols.insert(2, 'Overdose_Rate_X100K')

# Reorder the DataFrame based on the new column order
df_Supplemental_Data = df_Supplemental_Data[cols]
```

```python
#Import Census Financial data
df_Finance_2021 = pd.read_excel('CENSUS/est21all.xls')
df_Finance_2022 = pd.read_excel('CENSUS/est22all.xls')

# Concatenate all states into one
df_Finance = pd.concat([df_Finance_2021, df_Finance_2022])

#delete datasets that will no longer be used
del df_Finance_2021
del df_Finance_2022

#Capitalize all letters so all datasets match and can be joined correctly
df_Finance['county_name'] = df_Finance['county_name'].str.upper()
df_Finance['YEAR'] = df_Finance['YEAR'].astype(str)

df_Finance
```

```python
#Merge finance with the Final dataset
df_Supplemental_Data = pd.merge(df_Supplemental_Data, df_Finance, on=['county_name', 'ST_ABBREV', 'YEAR'], how='left' )
```
